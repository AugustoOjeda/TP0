### Introducción teórica


Una señal es cualquier magnitud medible que transporta información. Estas pueden ser unidimensionales, cuando la variable dependiente, que contiene la información a medir, se expresa en función de una sola variable independiente, o multidimensionales.
Asimismo, una señal puede ser continua en el tiempo, también llamada señal análoga, en la que la variable independiente toma valores reales continuos. Otro tipo es la señal discreta en el tiempo, también denominada señal digital, donde la variable independiente se encuentra cuantizada. Las señales discretas en el tiempo son esencialmente un arreglo de valore X[n] presentados como una función de un índice entero n. Estas pueden representarse gráficamente mediante técnicas computacionales o mediante funciones aritméticas y trigonométricas.
En el procesamiento digital de señales, una señal analógica es muestreada por un dispositivo de hardware, un convertidor analógico-digital, para formar una secuencia en tiempo discreto. Esta secuencia de entrada puede procesarse a través de un sistema DSP, generando una salida Y[n]. Finalmente, la salida es convertida nuevamente en una señal analógica y(t).
El muestreo de una señal señal analogica puede modelarse como la multiplicacion de la señal analogica en tiempo continuo por un tren de impulsos  períodicos con período T
$$
S(t) = \sum_{n=-\infty}^{\infty} \delta(t - nT)
$$
            El resultado de esta multiplicación es la señal de entrada muestreada $X_s(t)$ 
$$
X_s(t) = x(t)*s(t)
$$
    El muestreo analógico es una forma de modulación, donde la señal de entrada x(t) modula el tren de impulsos s(t) para formar $X_s(t)$, esta esta escalada por x(t) en los instantes donde ocurren los impulsos. Esta modulación, por tanto, “codifica” la amplitud de la señal analógica original en los momentos discretos 
$$
t=nT 
$$
    El teorema del muestreo estable que una señal continua x(t) puede ser reconstruida a partir de $X_s(t)$ siempre y cuando se cumpla que la señal sea limitada en banda y que el muestro se realize lo suficimente "rapido". Una señal en banda es aquella cuyo espectro $X(Ω)$ solo tiene energia en un rango finito de frecuencias:
$$
-Ω_b<Ω<Ω_b
$$
    Donde $Ω_b$ representa la mayor freciencia presente en $X(Ω)$. La convolucion de esta con $\delta(Ω-KΩ_s)$ produce traslaciones del espectro orginal a multiplos de la frecuencia de muestreo $KΩ_s$. Asi $X_s(Ω)$ está compuesta por réplicas del espectro de $X(Ω)$ centradas en multiplos de $Ω_s$.
    En cuanto a lo "rapido" existe una frecuencia minima o maxima de muestreo requerida para que sea posible reconstruir una señal a partir de sus muestras. Para evitar que la informacion original se mezcle con las réplicas de imagenes, se debe cumplir la condicion de $Ω_s>2Ω_b$.
    Es decir: La freciencia de muestreo debe ser al menos el doble de la frecuencia máxima de la señal, esta condicion se conoce como el criterio de muestreo de Nyquist, el cual estable la frecuencia minima de muestreo teórica para poder reconstruir una señal a partir de sus muestras. A su vez se consideran tres tipo de muestreos
##### 1.Sobre muestreo
Ocurre cuando $Ω_s>>2Ω_b$, en estos casos las replicas espectrales quedan muy separadas permitiendo una reconstrucción fácil y robusta
##### 2.Muestreo critico
Ocurre cuando $Ω_s=2Ω_b$, es caso limite del criterio de Nyquist, la replica de banda base toca justo a las réplicas de imagen en $Ω_s= \frac{ Ω_s }{2}$. En teoria estas recostrucciones son posibles pero se precisa de un filtro ideal para realizarse.
##### 3.Submuestreo
Ocurre cuando $Ω_s<2Ω_b$, produce superposicion de replicas tambien conocido como aliassing, la información original se mezca con las replicas y es imposible

Para analizar la importancia del criterio de Nyquist se empleó un generador de señales senoidales, al cual se le modificó la frecuencia hasta alcanzar condiciones de muestreo crítico y de submuestreo, configurando $N$ y el $F_s$ en 1000. El código fue adaptado conforme a los requerimientos establecidos por la cátedra, obteniéndose la siguiente versión. A esta implementación se le asignaron frecuencias de 1 Hz, 4 Hz, 150 Hz, 500 Hz, 990 Hz, 1001 Hz y 2001 Hz para su análisis. 

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Aug 13 18:36:22 2025

@author: Augusto
"""

import numpy as np
import matplotlib.pyplot as plt 
def mi_funcion_sen(vmax=1.0, dc=0.0, ff=1.0, ph=0.0, nn=1000, fs=1000.0, como_columna=True):
    tt = np.arange(nn, dtype=float) / float(fs)
    xx = dc + vmax * np.sin(2 * np.pi * ff * tt + ph)
    if como_columna:
        tt = tt.reshape(-1, 1)
        xx = xx.reshape(-1, 1)
    return tt, xx

if __name__ == "__main__":
    N = 1000
    fs = 1000

    casos = [(1, "1 Hz"),
             (4, "4 Hz"),
             (150, "150 Hz"),
             (500, "500 Hz"),
             (990, "990 Hz"),
             (1001, "1001 Hz"),
             (2001, "2001 Hz")]

    for ff, titulo in casos:
        tt, xx = mi_funcion_sen(vmax=1, dc=0, ff=ff, ph=0, nn=N, fs=fs, como_columna=False)
        plt.figure()
        plt.plot(tt, xx)
        plt.xlabel("tiempo [s]")
        plt.ylabel("Amplitud [V]")
        plt.title(f"Señal senoidal: {titulo}")
        plt.show()
```

Adicionalmente, se estudió el comportamiento de la función al emplear otras formas de señal, como en el caso de una señal triangular, obteniéndose los siguientes resultados: 


```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import signal
def mi_funcion_trian(vmax=1.0, dc=0.0, ff=1.0, ph=0.0, nn=1000, fs=1000.0, como_columna=True):
    tt = np.arange(nn, dtype=float) / float(fs)
    xx = dc + vmax * signal.sawtooth(2 * np.pi * ff * tt + ph)
    if como_columna:
        tt = tt.reshape(-1, 1)
        xx = xx.reshape(-1, 1)
    return tt, xx

if __name__ == "__main__":
    N = 1000
    fs = 1000

    casos = [(1, "1 Hz"),
             (4, "4 Hz"),
             (150, "150 Hz"),
             (500, "500 Hz"),
             (990, "990 Hz"),
             (1001, "1001 Hz"),
             (2001, "2001 Hz")]

    for ff, titulo in casos:
        tt, xx = mi_funcion_trian(vmax=1, dc=0, ff=ff, ph=0, nn=N, fs=fs, como_columna=False)
        plt.figure()
        plt.plot(tt, xx)
        plt.xlabel("tiempo [s]")
        plt.ylabel("Amplitud [V]")
        plt.title(f"Señal triangular: {titulo}")
        plt.show()
```

### Discusión y Conclusiones


Al comparar la Figura [1] con la Figura [2], se confirma que el gráfico obtenido es correcto y que ambos códigos presentan un funcionamiento adecuado, lo cual permite utilizarlos como base para extraer conclusiones. Por su parte, las Figuras [2] y [3] muestran señales representativas de senoidales. Esto se debe a que nos encontramos en la zona de sobremuestreo predicha por el criterio de Nyquist, con una reconstrucción de 1 ciclo y otra de 4 ciclos. En contraste, las figuras [4] y [9], evidencian que la señal ya no se reconoce claramente como senoidal. Esto ocurre porque nos encontramos en la región de aliasing, propia de los casos de submuestreo según el criterio de Nyquist. En el caso de la figura [5], tendriamos que tener un caso de muestreo critico ya que el valor de frecuencia es de 500 Hz, en vez de eso se obtiene un grafico donde no podemos obserbar una sinoideal pero no tengo una suficiente cantidad de puntos para reconstruir la onda. Ahora lo interensate ocurre en las Figuras [6], [7] y [8], ya que la teoria nos dice que en estos casos no tendriamos que ver ningun tipo de grafico parecido a una sinoideal, ya que el aliasing en esta zona tendria que ser enorme, en cambio podemos ver no solo que hay sinoideales bien marcadas si no que se repiten graficos como los de la figura [2], esto ocurre por que durante el muestreo, el espectro de la señal no solo aparece en la frecuencia original, sino que se replica de manera periódica en múltiplos de la frecuencia de muestreo (±fs, ±2fs, etc.). Al exceder el límite de Nyquist (fs/2), estas réplicas se solapan con el espectro original y producen aliasing. En consecuencia, una señal de alta frecuencia se manifiesta en el dominio temporal como una senoidal de frecuencia aparente más baja. La Figura [10],[11],[12],[13],[14],[15],[16] muestran que el mismo fenómeno se repite para señales triangulares. De este modo, se confirma que el aliasing no depende de la forma de onda empleada, sino exclusivamente de la relación entre la frecuencia de la señal y la frecuencia de muestreo. En consecuencia, tanto para señales senoidales como triangulares, se cumple lo predicho por la teoría del muestreo.


Como conclusión, los resultados obtenidos muestran que el comportamiento observado en las distintas señales coincide con lo predicho por el criterio de Nyquist. Se confirma que, independientemente de la forma de onda empleada, la correcta reconstrucción depende únicamente de la relación entre la frecuencia de la señal y la frecuencia de muestreo, verificándose así la teoría del muestreo y la aparición de aliasing en los casos de submuestreo.
